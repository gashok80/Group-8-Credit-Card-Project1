```python
# Initial imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_credit_card = pd.read_csv("C:/Users/gasho/Downloads/Daibetes/Resources/credit_card_transactions.csv")

# Display the DataFrame
df_credit_card['trans_date_trans_time']=pd.to_datetime(df_credit_card['trans_date_trans_time'])
df_credit_card
```

```python
# Check for missing values and data types
df_credit_card.info()

# Display missing values
df_credit_card.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of fraud and non-fraud transactions
plt.figure(figsize=(8, 6))
df_credit_card['is_fraud'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Fraudulent vs Non-Fraudulent Transactions')
plt.xlabel('Fraudulent')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Fraud', 'Fraud'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df_credit_card.groupby('category')['amt'].sum().sort_values().plot(kind='barh', color='coral')
plt.title('Total Transaction Amounts by Category')
plt.xlabel('Total Amount')
plt.ylabel('Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Convert transaction date and time to datetime format
df_credit_card['trans_date_trans_time'] = pd.to_datetime(df_credit_card['trans_date_trans_time'])

# Plot number of transactions over time
plt.figure(figsize=(12, 6))
df_credit_card.set_index('trans_date_trans_time').resample('M').size().plot()
plt.title('Number of Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.grid(True)
plt.show()
```

```python
# Plot gender distribution
plt.figure(figsize=(8, 6))
df_credit_card['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Genders')
plt.ylabel('')
plt.show()
```

```python
# Plot fraudulent transactions by merchant
plt.figure(figsize=(12, 8))
df_credit_card[df_credit_card['is_fraud'] == 1]['merchant'].value_counts().head(10).plot(kind='barh', color='salmon')
plt.title('Top 10 Merchants with Fraudulent Transactions')
plt.xlabel('Number of Transactions')
plt.ylabel('Merchant')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Add hour of the day to dataframe
df_credit_card['hour_of_day'] = df_credit_card['trans_date_trans_time'].dt.hour

# Plot transaction frequency by hour of the day
plt.figure(figsize=(10, 6))
df_credit_card['hour_of_day'].value_counts().sort_index().plot(kind='bar', color='darkblue')
plt.title('Transaction Frequency by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fraud_totals = df_credit_card.groupby('state')['is_fraud'].sum().reset_index()

# Sorting by amount in descending order
fraud_totals = fraud_totals.sort_values(by='is_fraud', ascending=False)

# Creating a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(y='state', x='is_fraud', data=fraud_totals)
plt.title('Total Fraud by State')
plt.xlabel('Total Fraud')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# Box plot of transaction amounts
plt.figure(figsize=(10, 6))
df_credit_card.boxplot(column='amt', by='is_fraud', grid=False, color=dict(boxes='salmon', whiskers='blue', medians='red', caps='black'))
plt.title('Box Plot of Transaction Amounts')
plt.suptitle('')
plt.xlabel('Fraudulent')
plt.ylabel('Amount')
plt.show()
```

```python
# Summary of findings
fraud_summary = df_credit_card.groupby('is_fraud').agg({
    'amt': ['mean', 'median', 'std', 'sum'],
    'category': 'count',
    'merchant': 'count'
}).reset_index()

fraud_summary.columns = ['Fraudulent', 'Average Amount', 'Median Amount', 'Amount Std Dev', 'Total Amount', 'Category Count', 'Merchant Count']
fraud_summary
```
